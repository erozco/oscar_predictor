```python
import numpy as np
import pandas as pd
```

```python
# let's do some simple preprocessing
movies_df = pd.read_csv('../data/movie_data_2000_2020.csv')
movies_df

# let's narrow it to oscar best picture nominees
nominated_df = movies_df.loc[movies_df['Oscar_Best_Picture_nominated'] == 'Yes']

# we are trying to predicting best picture winner

target = nominated_df['Oscar_Best_Picture_won']
target.to_csv('../data/best_pic_win.csv')

# given that we don't have this data for 2022's nominees, let's drop what we don't need

known_df = nominated_df[['year','movie','certificate','duration','genre','rate','metascore','votes',\
                    'Oscar_nominated','Oscar_nominated_categories','release_date.year',\
                     'release_date.month','release_date.day-of-week']]
known_df.to_csv('../data/known.csv')
```
